```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#load dataset
assembly_df = pd.read_csv('/content/assembly_cons.csv')
parliamentary_df = pd.read_csv('/content/party_wise.csv')
state_wise_df = pd.read_csv('/content/state_wise.csv')
by_election_df = pd.read_csv('/content/by_elec.csv')
```

```python
import matplotlib.pyplot as plt
bjp_wins = assembly_df[assembly_df['party'].str.contains('Bharatiya Janata Party')]
congress_wins = assembly_df[assembly_df['party'].str.contains('Indian National Congress')]

# Calculate total wins
total_bjp_wins = bjp_wins['won'].sum()
total_congress_wins = congress_wins['won'].sum()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(['BJP'], [total_bjp_wins], color='orange', label='BJP')
plt.bar(['Congress'], [total_congress_wins], color='blue', label='Congress')
plt.title('Comparison of Wins: BJP vs Congress in Assembly Elections')
plt.xlabel('Party')
plt.ylabel('Total Wins')
plt.legend()
plt.show()
```

```python
import seaborn as sns

# Assuming 'state_wise_df' contains state-wise data for different parties
top_states = state_wise_df.nlargest(5, 'won')

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='state', y='won', hue='party', data=top_states, palette='Set2')
plt.title('Top 5 States with Highest BJP Wins in Assembly Elections')
plt.xlabel('State')
plt.ylabel('Total Wins')
plt.xticks(rotation=45)
plt.legend(title='Party')
plt.show()

```

```python
import numpy as np

# Aggregate wins by party and state
state_party_wins = state_wise_df.groupby(['state', 'party'])['won'].sum().unstack()

# Plotting
plt.figure(figsize=(14, 10))
state_party_wins.plot(kind='bar', stacked=True, colormap='Set2', figsize=(20, 30))
plt.title('State-wise Distribution of Party Wins in Assembly Elections')
plt.xlabel('State')
plt.ylabel('Total Wins')
plt.xticks(rotation=45)
plt.legend(title='Party')
plt.show()
```

```python
import seaborn as sns

# Assuming 'parliamentary_df' contains data for wins in parliamentary constituencies
plt.figure(figsize=(10, 8))
sns.barplot(x='Won', y='Party', data=parliamentary_df, hue='Party', palette='Set2', ci=None)
plt.title('Comparison of Wins between BJP and Congress in Parliamentary Constituencies')
plt.xlabel('Total Wins')
plt.ylabel('Party')
plt.legend(title='Party', loc='upper right')
plt.show()
```
